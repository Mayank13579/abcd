```python
def detect_outliers(data):
    Q1=data.quantile(0.25)
    Q3=data.quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    outlier = data[(data < lower_bound) | (data > upper_bound)]
    return outlier
```

```python

outlier = detect_outliers(data['sepal width (cm)'])
```

```python
data_wo_outlier = data[~data['sepal width (cm)'].isin(outlier)]
```

```python
sns.boxplot(data_wo_outlier, x = 'sepal width (cm)', color = 'Blue')
```
